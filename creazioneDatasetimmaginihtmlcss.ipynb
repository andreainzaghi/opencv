```python
from bs4 import BeautifulSoup
import json

data = {}
data['tag'] = []

# Carica il codice HTML da analizzare
html = open('./codiceHTML.html').read()

# Analizza il codice HTML utilizzando BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')

# Cerca il tag nav nella pagina HTML
tag = soup.findAll('nav')

# cicla gli elementi trovati e salvati inn tag e estrai la classe
csssclass=''
for tag1 in tag:
    cls = tag1.get('class')
    if cls is None:
        cls = tag1.get('id')
        if cls is not None:
            #verifica n clel file codiceCSS.css se la classe esiste e stampa il risultato
            css = open('./codiceCSS.css').read()
            if cls[0] in css:

                #stampa tutta la classe css tra le parentesi graffe
                csssclass=css[css.find(cls[0]):css.find('}',css.find(cls[0]))+1]
        
            #prensi il tag nav e la relativa classe e aggiungerli e visualizzalo in un json 
            data['tag'].append({
                'tag': tag1,
            
                'css':csssclass
            })

    # Stampa il risultato in formato JSON
            print(data)


        else:
            print('La classe non è presente nel file css')






# Stampa la classe del tag nav

```
